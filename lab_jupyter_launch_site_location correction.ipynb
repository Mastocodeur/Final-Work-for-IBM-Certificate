Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `launch_sites` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
# Function to determine marker color based on launch class
def get_marker_color(launch_class):
    if launch_class == 1:
        return 'green'
    else:
        return 'red'
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

# Create a new column 'marker_color' in launch_sites_df
spacex_df['marker_color'] = spacex_df['class'].apply(get_marker_color)

# Add markers to the MarkerCluster for each launch record
for index, record in spacex_df.iterrows():
    # Get coordinates and marker color
    coordinate = [record['Lat'], record['Long']]
    marker_color = record['marker_color']
    
    # Create Marker with color based on launch class
    marker = folium.Marker(
        location=coordinate,
        icon=folium.Icon(color=marker_color),
        popup=f"Launch Site: {record['Launch Site']}, Class: {record['class']}"
    )
    
    # Add the marker to the MarkerCluster
    marker_cluster.add_child(marker)

# Add the MarkerCluster to the site_map
site_map.add_child(marker_cluster)

# Display the updated map with clustered markers
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import radians, sin, cos, sqrt, atan2

# Function to calculate distance between two points using Haversine formula
def haversine(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    # Radius of Earth in kilometers
    R = 6371.0
    
    # Calculate the distance
    distance = R * c
    return distance

# Function to calculate distances from a launch site to all proximities
def calculate_distances(launch_site, proximities_df):
    distances = []
    for index, proximity in proximities_df.iterrows():
        distance = haversine(
            launch_site['Lat'],
            launch_site['Long'],
            proximity['Lat'],
            proximity['Long']
        )
        distances.append(distance)
    return distances

# List to store distances for each launch site
launch_distances = []

# Create a DataFrame to store proximities
proximities_df = pd.DataFrame({
    'Lat': launch_sites_df['Lat'],
    'Long': launch_sites_df['Long'],
    'Launch Site': launch_sites_df['Launch Site']
})

# Loop through each launch site and calculate distances to proximities
for index, site in launch_sites_df.iterrows():
    site_name = site['Launch Site']
    distances = calculate_distances(site, proximities_df)
    
    # Create a dictionary to store site name and distances
    site_distances = {
        'Launch Site': site_name,
        'Distances': distances
    }
    
    # Append the dictionary to launch_distances list
    launch_distances.append(site_distances)

# Convert launch_distances to a DataFrame for easier analysis
launch_distances_df = pd.DataFrame(launch_distances)

# Print the DataFrame showing launch sites and their distances to proximities
print(launch_distances_df)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Define the coordinates of the coastline
coastline_lat = 28.56367
coastline_lon = -80.57163

# Function to calculate distance between a launch site and the coastline
def calculate_distance_to_coastline(launch_site_lat, launch_site_lon):
    return haversine(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Loop through each launch site to find the closest coastline
for index, site in launch_sites_df.iterrows():
    launch_site_name = site['Launch Site']
    launch_site_lat = site['Lat']
    launch_site_lon = site['Long']
    
    # Calculate distance to the coastline
    distance_coastline = calculate_distance_to_coastline(launch_site_lat, launch_site_lon)
    
    # Create a Marker for the closest coastline point
    closest_coastline_marker = folium.Marker(
        [coastline_lat, coastline_lon],
        icon=folium.DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
        ),
        popup='Closest Coastline to ' + launch_site_name
    )
    
    # Add the closest coastline marker to the site_map
    site_map.add_child(closest_coastline_marker)

# Display the updated map with closest coastline markers
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
```

```python
# Define the coordinates of the points of interest (example coordinates)
points_of_interest = {
    'City': {'Lat': 29.7604, 'Lon': -95.3698},  # Example city coordinates (Houston)
    'Railway': {'Lat': 29.7243, 'Lon': -95.4018},  # Example railway coordinates
    'Highway': {'Lat': 29.7105, 'Lon': -95.3989}  # Example highway coordinates
}

# Function to calculate distance between a launch site and a point of interest
def calculate_distance_to_point_of_interest(launch_site_lat, launch_site_lon, poi_lat, poi_lon):
    return haversine(launch_site_lat, launch_site_lon, poi_lat, poi_lon)

# Loop through each launch site to find the distances to points of interest
for index, site in launch_sites_df.iterrows():
    launch_site_name = site['Launch Site']
    launch_site_lat = site['Lat']
    launch_site_lon = site['Long']
    
    # Create empty dictionary to store distances to points of interest
    distances_to_poi = {}
    
    # Loop through each point of interest
    for poi_name, poi_coords in points_of_interest.items():
        poi_lat = poi_coords['Lat']
        poi_lon = poi_coords['Lon']
        
        # Calculate distance to the point of interest
        distance_poi = calculate_distance_to_point_of_interest(launch_site_lat, launch_site_lon, poi_lat, poi_lon)
        
        # Store the distance in the dictionary
        distances_to_poi[poi_name] = distance_poi
    
    # Get the closest point of interest and its distance
    closest_poi = min(distances_to_poi, key=distances_to_poi.get)
    closest_poi_distance = distances_to_poi[closest_poi]
    
    # Create a Marker for the closest point of interest
    closest_poi_marker = folium.Marker(
        [points_of_interest[closest_poi]['Lat'], points_of_interest[closest_poi]['Lon']],
        icon=folium.DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(closest_poi_distance),
        ),
        popup='Closest ' + closest_poi + ' to ' + launch_site_name
    )
    
    # Create a PolyLine between launch site and closest point of interest
    polyline_poi = create_polyline([launch_site_lat, launch_site_lon], [points_of_interest[closest_poi]['Lat'], points_of_interest[closest_poi]['Lon']])
    
    # Add the Marker and PolyLine to the site_map
    site_map.add_child(closest_poi_marker)
    site_map.add_child(polyline_poi)

# Display the updated map with markers and lines to points of interest
site_map
```
